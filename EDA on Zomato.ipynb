```python
#we are dropping the uneccesary columns 
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

DROPPING DUPLICATE VALUES

```python
df.drop_duplicates(inplace=True)
df.shape
```

CLEANING RATE COLUMN

```python
#we simply want the & String values rather than /5
df['rate'].unique()
```

Removing "NEW" , "-" and "/5" from Rate Column

```python
def handlerate(value):
    if(value=='NEW'or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)

df['rate']=df['rate'].apply(handlerate)
df['rate']
```

```python
df.rate.isnull().sum()
```

Filling Null Values in Rate Column with Mean

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
##Dropping Null Values
df.dropna(inplace=True)
df
```

```python
#rename the lengthy columns
df.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace=True)
df
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df
```

```python
df['Cost2plates'].unique()
```

```python
#removing in between commas from every value
def handlecomma(value):
    value=str(value)
    if',' in value:
        value=value.replace(',','')
    else:
        return float(value) #if comma is not present in the value then just return the value as it is

df['Cost2plates']=df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df
```

```python
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
#now we group the rest_type<1000 and make them cluster as others
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
#now we make all the above rest_types as others
def handle_rest_type(value):
    if( value in rest_types_lessthan1000):
        return'others'
    else:
        return value

df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
#others= 9003
```

```python
df.head()
```

```python
#till now we cleaned name,online_order,book_table,rate,rest_type,Cost2plates
```

```python
#now we clean Location Column
#we apply same as above for Location Column
```

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
location_lessthan300=location[location<300] # we took less than 300 coz as highest value is about 5000 so we did the average of it and then took 300

def handle_location(value):
   if(value in location_lessthan300):
       return'others'
   else:
       return value

df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
#others=> 4954
```

```python
df
```

```python
# now we clean Cuisines column
```

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisines_lessthan100=cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
#others=>2852
df.head()
```
